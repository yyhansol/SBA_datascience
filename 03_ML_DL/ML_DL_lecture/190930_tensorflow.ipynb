```
mnist = tf.keras.datasets.mnist
mnist
```

```
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```
X_train[0]
```

```
y_train[0:3]
```

```
X_train, X_test= X_train/255.0, X_test/255.0 ## 정규화
```

```
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),  #입력노드 개수 28*28
    tf.keras.layers.Dense(512, activation=tf.nn.relu),  #은닉층 노드 개수 512개, 활성함수 relu
    tf.keras.layers.Dropout(0,2),  
    tf.keras.layers.Dense(10, activation=tf.nn.softmax)  #결과층 노드 개수 10개, 활성함수 softmax
])
```

```
model.compile(optimizer='adam',   #최적화
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```
model.fit(X_train, y_train, epochs=5)  #5번 학습
```

```
model.evaluate(X_test, y_test)
```
