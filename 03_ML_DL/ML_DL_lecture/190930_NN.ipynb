```
import numpy as np
import matplotlib.pyplot as plt
```

```
x = np.linspace(-5, 5, 300)

step_fnc = x.copy()
step_fnc[step_fnc <= 0] = 0
step_fnc[step_fnc>0] = 1

sig_fnc = 1/(1+np.exp(-x))

relu_fnc = np.maximum(x,0)

tanh_fnc = np.tanh(x)

```

```
import matplotlib
matplotlib.rcParams['axes.unicode_minus'] = False
```

```
import matplotlib.pylab as plt
plt.rcParams["figure.figsize"] = (14,10)
```

```
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2)
ax1.plot(x, step_fnc, label='step',color='orange')
ax2.plot(x, sig_fnc, label='sigmoid',color='green')
ax3.plot(x, relu_fnc, label='relu',color='red')
ax4.plot(x, tanh_fnc, label='tanh')
```

```
plt.plot(x, tanh_fnc, label='tanh')
plt.plot(x, relu_fnc, label='relu')
plt.plot(x, sig_fnc, label='sigmoid')
plt.plot(x, step_fnc, label='step function')
plt.legend(loc='best')
plt.xlabel('x')
plt.ylabel('relu(x), tanh(x)')
```
