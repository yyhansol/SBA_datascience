## 3. 모델 만들기
train -> 800개 , test -> 294개

층 4개( 3X3 convolution, 2X2 pooling, 활성화 함수는 relu)

첫 번째 층 필터 32개

두 번째 층 필터 64개

세 번째 층 필터 128개

네 번째 층 노드 256개(fully connect layer)


```python
X = tf.placeholder(tf.float32, [None, 32, 32, 3])  
Y = tf.placeholder(tf.float32, [None,2])
keep_prob = tf.placeholder(tf.float32)
```

```python
W1 = tf.Variable(tf.random_normal([3,3,3,32], stddev=0.01))
L1 = tf.nn.conv2d(X, W1, strides=[1,1,1,1], padding='SAME')
L1 = tf.nn.relu(L1)
L1 = tf.nn.max_pool(L1, ksize=[1,2,2,1], strides=[1,2,2,1], padding='SAME')
L1 = tf.nn.dropout(L1, keep_prob)
print(L1)
```

```python
W2 = tf.Variable(tf.random_normal([3,3,32,64], stddev=0.01))
L2 = tf.nn.conv2d(L1, W2, strides=[1,1,1,1], padding='SAME')
L2 = tf.nn.relu(L2)
L2 = tf.nn.max_pool(L2, ksize=[1,2,2,1], strides=[1,2,2,1], padding='SAME')
L2 = tf.nn.dropout(L2, keep_prob)

print(L2)
```

```python
W3 = tf.Variable(tf.random_normal([3,3,64, 128], stddev=0.01))
L3 = tf.nn.conv2d(L2, W3, strides=[1,1,1,1], padding='SAME')
L3 = tf.nn.relu(L3)
L3 = tf.nn.max_pool(L3, ksize=[1,2,2,1], strides=[1,2,2,1], padding='SAME')
L3 = tf.nn.dropout(L3, keep_prob)

print(L3)
```

```python
W4 = tf.Variable(tf.random_normal([4 * 4 * 128, 256], stddev=0.01))
L4 = tf.reshape(L3, [-1, 4 *4 * 128])
L4 = tf.matmul(L4, W4)
L4 = tf.nn.relu(L4)
print(L4)
```

```python
W5 = tf.Variable(tf.random_normal([256,2], stddev=0.01))
model = tf.matmul(L4, W5)
model
```

```python
cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits=model, labels=Y))
optimizer = tf.train.AdamOptimizer(0.001).minimize(cost)
```

```python
init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)
```

```python
batch_size = 100
total_batch = int(X_train.shape[0]/ batch_size)
total_batch
epochs = 20
```

```python
def next_batch(start, num, data, labels):
    data_X = data[start:start+num]
    data_y = labels[start:start+num]
    return np.asarray(data_X), np.asarray(data_y)
```

```python
print(X_train.shape, Y_train.shape)
batch_X, batch_y = next_batch(0,10, X_train, Y_train)
print(batch_X.shape, batch_y.shape)
```

## 4. 학습시키기
만든 모델을 train 이미지로 에폭 수 20, 과대적합을 막기위해 dropout rate을 0.5로 설정 후 학습을 시켜주었습니다. 학습시킨 후에는 test 이미지로 dropout rate를 1로 설정하여 정확도를 측정해주었습니다.

```python
total_cost = 0
epoch = 0
print("전체 입력 데이터 : {}".format(X_train.shape))
print("전체 출력 데이터 : {}".format(Y_train.shape))

for epoch in range(20):
    total_cost = 0
    for i in range(total_batch):
        batch_xs, batch_ys = next_batch(batch_size*i, batch_size, X_train, Y_train)
        batch_xs = batch_xs.reshape(-1, 32, 32, 3)
        _, cost_val = sess.run([optimizer, cost], feed_dict={X: batch_xs, Y: batch_ys, keep_prob:0.8})
        total_cost += cost_val
        if (i==0 or i==total_batch-1):
            print('data_step = {}, Avg. cost = {:.3f}'.format(i, cost_val))
    print('epoch: {} total.cost = {:.3f}'.format(epoch, total_cost))
```

```python
print(X_test.shape, Y_test.shape)
```

```python
is_correct = tf.equal(tf.argmax(model,1), tf.argmax(Y,1))
accuracy = tf.reduce_mean(tf.cast(is_correct, tf.float32))
print('정확도 : ', sess.run(accuracy,
                        feed_dict={X:X_test.reshape(-1,32,32,3),
                                  Y:Y_test,
                                  keep_prob:0.8}))
```

## 6. 결론
각 레이블 당 600개 밖에 되지 않는 이미지와 4개의 층을 갖는 모델, 적은 학습 에폭수로도 높은 정확도를 낼 수 있었던 이유는 다음과 같이 생각하였습니다. 첫번 째, 익은 토마토와 익지 않은 토마토의 색깔이 분명하게 차이가 나기 때문에 학습하기 쉬웠을 것으로 예상했습니다. 두 번째는 토마토는 원의 형태를 갖고 있어서 각도에 따라 그 모양이 크게 변하지 않아 다른 사물에 비해 학습이 쉬웠을 것입니다. 이 프로젝트를 바탕으로 더 나아가서 숙성의 정도에 따라 더 세분화된 레이블로 나누어 각 숙성도마다 재배의 시기를 예측할 수 있는 모델을 만들어보면 의미가 있을 것이라고 생각했습니다.
