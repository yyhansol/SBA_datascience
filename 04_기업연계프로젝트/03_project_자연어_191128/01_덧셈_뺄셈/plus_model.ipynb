```python
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
```

```python
def elbow(X):
    sse=[]
    for i in range(1,11):
        km=KMeans(n_clusters=i,init="k-means++",random_state=0)
        km.fit(X)
        sse.append(km.inertia_)
    plt.plot(range(1, 11),sse,marker="o")
    plt.show()
```
