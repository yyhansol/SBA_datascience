```python
text = """
    김찬영 020822-3069110
    김준영 040825-3069115
    김채영 110715-4063111
"""
```

```python
pattern = re.compile('\d{6}[ -.]?\d{7}')
```

```python
pattern.findall(text)
```

```python
pattern = re.compile('(?P<birth>\d{6})-(?P<secret>\d{7})')
```

```python
pattern.findall(text)
```

```python
result = pattern.sub('\g<birth>-*******', text)
```

```python
print(result)
```

```python
pattern = re.compile('(?P<name>\w{3}) (?P<birth>\d{6})-(?P<secret>\d{7})')
```
